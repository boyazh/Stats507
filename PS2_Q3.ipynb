```python
#Q1.1
#PS2 Question 3a
import pandas as pd
import numpy as np


df1 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/DEMO_J.XPT')
df2 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/DEMO_I.XPT')
df3 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/DEMO_H.XPT')
df4 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/DEMO_G.XPT')

df1['source'] = '17-18'
df2['source'] = '15-16'
df3['source'] = '13-14'
df4['source'] = '11-12'

DEMO_df = df1.append(df2).append(df3).append(df4)

DEMO_df = DEMO_df.loc[:, ['SEQN','RIDAGEYR',
                          'RIDRETH3','DMDEDUC2',
                          'DMDMARTL', 'RIDSTATR', 
                          'SDMVPSU', 'SDMVSTRA', 
                          'WTMEC2YR', 'WTINT2YR',
                          'source']]
DEMO_df = DEMO_df.rename(columns = {'SEQN':'unique ids',
                                    'RIAGENDR': 'gender',
                                    'RIDAGEYR':'age',
                                    'RIDRETH3':'race and ethnicity',
                                    'DMDEDUC2':'education', 
                                    'DMDMARTL':'marital status'})

#DEMO_df = DEMO_df.replace(to_replace = np.nan, value = -1)

DEMO_df['race and ethnicity'].replace(to_replace = [1, 2, 3, 4, 6, 7, np.nan], 
                                      value = ['mexican american','other hispanic',
                                               'non-hispanic white', 'non-hispanic black',
                                               'non-hispanic asian', 'other race', 'missing'], inplace = True)
DEMO_df['education'].replace(to_replace = [1, 2, 3, 4, 5, 7, 9, np.nan], 
                             value = ['less than 9th grade', '9-11th grade','high school graduate',
                                        'some college or AA degree','clooege graduate or above', 
                                      'refused','dont know', 'missing'], inplace = True)
DEMO_df['marital status'].replace(to_replace = [1, 2, 3, 4, 5, 6, 77, 99, np.nan], 
                                 value = ['married', 'widowed', 'divorced', 'separated', 'never married', 
                                          'living with partner', 'refused', 'dont know', 'missing'], inplace = True)

DEMO_df.to_pickle('DEMO_df.pickle')



#PS2 Question 3b
import pandas as pd
import numpy as np

df5 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/OHXDEN_J.XPT')
df6 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2015-2016/OHXDEN_I.XPT')
df7 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2013-2014/OHXDEN_H.XPT')
df8 = pd.read_sas('https://wwwn.cdc.gov/Nchs/Nhanes/2011-2012/OHXDEN_G.XPT')


df5['source'] = '17-18'
df6['source'] = '15-16'
df7['source'] = '13-14'
df8['source'] = '11-12'

DEN_df = df5.append(df6).append(df7).append(df8)

b = ['SEQN','OHDDESTS']
for i in range(1,33): 
    if i < 10:
        a = 'OHX0{}TC'.format(i)
        DEN_df[a].replace(to_replace = [1, 2, 3, 4, 5, 9, np.nan], 
                           value = ['primary tooth present', 'permanent tooth present', 
                                   'dental implant', 'tooth not present', 
                                    'permant dental root fragment present', 
                                    'could not assess', 'missing'], inplace = True)
        b_ = b.append(a)
    else:
        a = 'OHX{}TC'.format(i)
        DEN_df[a].replace(to_replace = [1, 2, 3, 4, 5, 9, np.nan], 
                           value = ['primary tooth present', 'permanent tooth present', 
                                   'dental implant', 'tooth not present', 
                                    'permant dental root fragment present', 
                                    'could not assess', 'missing'], inplace = True)
        b_ = b.append(a)

c = []
for j in range(2,32):
    if j < 10:
        a = 'OHX0{}CTC'.format(j)
        c_ = c.append(a)
    elif 10 <= j <16:
        a = 'OHX{}CTC'.format(j)
        c_ = c.append(a)
    elif j > 17:
        a = 'OHX{}CTC'.format(j)
        c_ = c.append(a)
        
c_ = c.append('source')

lis_= b + c

DEN_df = DEN_df.loc[:,lis_]

DEN_df.to_pickle('DEN_df.pickle')


#PS2 question 3c

DEMO_cases = pd.DataFrame(DEMO_df.count())

DEN_cases = pd.DataFrame(DEN_df.count())

print(DEMO_cases)
print(DEN_cases)
# -

```
